# Feature scaling

```python
from sklearn.model_selection import train_test_split
x=nw_df.iloc[:,:-1]
x
```

```python
y=nw_df.iloc[:,-1]
y
```

```python
x.shape
```

```python
y.shape
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.33,random_state=43)
```

```python
train_x.shape
```

```python
test_x.shape
```

# Model building

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
lm=LinearRegression()
lm.fit(train_x,train_y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
lm.score(train_x,train_y)
```

```python
pred=lm.predict(test_x)
print("predicted_score",pred)
print("actual_score",test_y)
```

```python
print('error:')
print('mean_abosolute_error',mean_absolute_error(test_y,pred))
print('mean_squared_error',mean_squared_error(test_y,pred))
print('root_mean_squared_error',np.sqrt(mean_squared_error(test_y,pred)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(test_y,pred))
```

```python
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=.0001)
ls.fit(train_x,train_y)
predl=ls.predict(test_x)
print(ls.score(train_x,train_y))
print("mean squared error:",np.sqrt(mean_squared_error(test_y,predl)))
mean_squared_error(test_y,predl)
```

```python
rg=Ridge(alpha=0.000001)
rg.fit(train_x,train_y)
print(rg.score(train_x,train_y))
predr=rg.predict(test_x)
print(mean_squared_error(test_y,predr))
print("mean squred error:",np.sqrt(mean_squared_error(test_y,predr)))
```

# hypertuning

```python
from sklearn.model_selection import GridSearchCV
alpha_val={'alpha':[0.000001,1,0.1,.001,.0001,.00001,.5,0]}
model=Ridge()
gd_sr=GridSearchCV(estimator=model,param_grid=alpha_val)
gd_sr.fit(train_x,train_y)
best_parameter=gd_sr.best_params_
print(best_parameter)
best_score=gd_sr.best_score_
print(best_score)
```

```python
from sklearn.model_selection import GridSearchCV
alpha_val={'alpha':[0.000001,1,0.1,.001,.0001,.00001,.5,0]}
model=Lasso()
gd_sr=GridSearchCV(estimator=model,param_grid=alpha_val)
gd_sr.fit(train_x,train_y)
best_parameter=gd_sr.best_params_
print(best_parameter)
best_score=gd_sr.best_score_
print(best_score)
```

```python
import xgboost as xgb
ad=xgb.XGBRegressor()
grid={'n_estimators':[50,100,200,300]}
gd_sr=GridSearchCV(estimator=ad,param_grid=grid)
gd_sr.fit(train_x,train_y)
best_parameter=gd_sr.best_params_
print(best_parameter)
best_score=gd_sr.best_score_
print(best_score)
```
